### Waveform visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set() # Use seaborn's default style

# Load your sound file
snd = parselmouth.Sound("your_audio_file.wav")

plt.figure()
plt.plot(snd.xs(), snd.values.T)
plt.xlim([snd.xmin, snd.xmax])
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.show()
```

### Spectrogram

```python
spectrogram = snd.to_spectrogram()
print(f"Shape of X: {spectrogram.x_grid().shape}")
print(f"Shape of Y: {spectrogram.y_grid().shape}")
print(f"Shape of C: {spectrogram.values.T.shape}")

plt.figure()
plt.pcolormesh(spectrogram.x_grid(), spectrogram.y_grid(), spectrogram.values.T)
plt.xlabel("Time [s]")
plt.ylabel("Frequency [Hz]")
plt.title("Spectrogram")
plt.colorbar(label="Intensity [dB]")
plt.xlim([spectrogram.xmin, spectrogram.xmax])
plt.ylim([spectrogram.ymin, spectrogram.ymax])
plt.show()
```

### Pitch contour

```python
pitch = snd.to_pitch()

plt.figure()
plt.plot(pitch.xs(), pitch.selected_array['frequency'])
plt.xlim([snd.xmin, snd.xmax])
plt.xlabel("Time [s]")
plt.ylabel("Frequency [Hz]")
plt.title("Pitch Contour")
plt.ylim([75, 300]) # Adjust y-axis limits as needed
plt.show()
```
